```python
#import csv และ networkx
import csv
from operator import itemgetter
import networkx as nx
from networkx.algorithms import community
```

```python
#นำเข้าข้อมูล
with open('fb_target_edit.csv', 'r') as nodecsv: # Open the file
    nodereader = csv.reader(nodecsv) # Read the csv
    # Retrieve the data 
    nodes = [n for n in nodereader][1:]

node_names = [n[0] for n in nodes] # Get a list of only the node names

with open('fb_edges_edit.csv', 'r') as edgecsv: # Open the file
    edgereader = csv.reader(edgecsv) # Read the csv
    edges = [tuple(e) for e in edgereader][1:] # Retrieve the data
```

```python
#สำรวจโหนด
print(len(node_names))
```

```python
#สำรวจจุดเชื่อม (edges)
print(len(edges))
```

```python
#ทำการสร้างกราฟ
G = nx.Graph()
G.add_nodes_from(node_names)
G.add_edges_from(edges)
```

```python
#ดูข้อมูลของกราฟ
print(nx.info(G))
```

```python
#สร้าง attributes ในฟอร์มของ dictionary
facebook_id_dict = {}
page_name_dict = {}
page_type_dict = {}
```

```python
#ตั้งค่า loop ใน Node list 
for node in nodes:
    facebook_id_dict[node[0]] = node[1]
    page_name_dict[node[0]] = node[2]
    page_type_dict[node[0]] = node[3]
```

```python
#ตั้งค่า node attributes
nx.set_node_attributes(G, facebook_id_dict, 'Facebook_ID')
nx.set_node_attributes(G, page_name_dict, 'page_name')
nx.set_node_attributes(G, page_type_dict, 'page_type')
```

```python
#ตรวจสอบว่าสามารถดูข้อมูลโหนดของ Facebook ID ได้หรือไม่
for n in G.nodes(): 
    print(n, G.nodes[n]['Facebook_ID'])
```

```python
#ตรวจสอบว่าสามารถดูข้อมูลโหนดของ Page name ได้หรือไม่
for n in G.nodes(): 
    print(n, G.nodes[n]['page_name'])
```

```python
#ตรวจสอบว่าสามารถดูข้อมูลโหนดของ Page type ได้หรือไม่
for n in G.nodes(): 
    print(n, G.nodes[n]['page_type'])
```

```python
#ดูค่า network density
density = nx.density(G)
print("Network density:", density)
```

```python
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
```

```python
#สร้างการเชื่อมต่อ betweenness และ eigenvector
betweenness_dict = nx.betweenness_centrality(G) # Run betweenness centrality
eigenvector_dict = nx.eigenvector_centrality(G) # Run eigenvector centrality
```

```python
nx.set_node_attributes(G, betweenness_dict, 'betweenness')
nx.set_node_attributes(G, eigenvector_dict, 'eigenvector')
```

```python
#ตรวจสอบโหนดที่มีการเชื่อม betweenness centrality มากที่สุด 20 โหนด
sorted_betweenness = sorted(betweenness_dict.items(), key=itemgetter(1), reverse=True)

print("Top 20 nodes by betweenness centrality:")
for b in sorted_betweenness[:20]:
    print(b)
```

```python
#First get the top 20 nodes by betweenness as a list
top_betweenness = sorted_betweenness[:20]

#Then find and print their degree
for tb in top_betweenness: # Loop through top_betweenness
    degree = degree_dict[tb[0]] # Use degree_dict to access a node's degree, see footnote 2
    print("Name:", tb[0], "| Betweenness Centrality:", tb[1], "| Degree:", degree)
```

```python
communities = community.greedy_modularity_communities(G)
```

```python
modularity_dict = {} # Create a blank dictionary
for i,c in enumerate(communities): # Loop through the list of communities, keeping track of the number for the community
    for name in c: # Loop through each person in a community
        modularity_dict[name] = i # Create an entry in the dictionary for the person, where the value is which group they belong to.

# Now you can add modularity information like we did the other metrics
nx.set_node_attributes(G, modularity_dict, 'modularity')
```

```python
for i,c in enumerate(communities): # Loop through the list of communities
    if len(c) > 2: # Filter out modularity classes with 2 or fewer nodes
        print('Class '+str(i)+':', list(c)) # Print out the classes and their members
```
